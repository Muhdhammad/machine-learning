```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area, df.price)
```

```python
reg = linear_model.LinearRegression()      # obj of LinearRegression
reg = reg.fit(df[['area']],df.price)               # fit to train the model
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
135.78767123 * 3300 + 180616.43835616432
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area, df.price, color='red')
plt.plot(df.area, reg.predict(df[['area']]))
```

```python
d = pd.read_csv('areas.csv')
d.head()
```

```python
p = reg.predict(d)
```

```python
d['price'] = p
```
