```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
```

```python
image = imread('flower.jpg')
```

```python
image.shape    # height, width, color channels RGB
```

```python
plt.imshow(image)
```

```python
X = image.reshape(-1, 3)
```

```python
X.shape
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=8).fit(X)
```

```python
kmeans.labels_   # label of each point, belongs to which cluster
```

```python
kmeans.cluster_centers_    # centriod of each cluster 
```

```python
kmeans.inertia_
```

```python
kmeans.n_iter_
```

```python
segmented_image = kmeans.cluster_centers_[kmeans.labels_]
```

```python
# reshaping into the dimensions of original image

segmented_image = segmented_image.reshape(image.shape)
```

```python
plt.imshow(segmented_image/255)
```
