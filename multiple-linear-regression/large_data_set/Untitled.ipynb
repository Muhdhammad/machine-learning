```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('BostonHousing.csv')
df
```

```python
df.head()
```

```python
# Splitting dataset into x and y variables
```

```python
y = df.medv
y
```

```python
x = df.drop(['medv'], axis=1)
x
```

```python
# Data Split
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Perform 80/20 data split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2) # 80% for training, 20$ for testing
```

```python
# Data Dimension
```

```python
x_train.shape, y_train.shape   
```

```python
x_test.shape, y_test.shape
```

```python
# linear regression model
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
# Build training model
```

```python
model.fit(x_train, y_train)
```

```python
# apply trained model to predict based on test data
```

```python
y_predict = model.predict(x_test)
```

```python
y_predict
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_test
```

```python
y_predict
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.scatter(y_test,y_predict, marker='+', alpha=0.3)
```

```python
# Evaluation
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_test, y_predict)
mse
```

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(y_test, y_predict)
print(f'R^2 Score: {r2 * 100:.2f}%')
```

```python
train_score = model.score(x_train, y_train)
test_score = model.score(x_test, y_test)
```
